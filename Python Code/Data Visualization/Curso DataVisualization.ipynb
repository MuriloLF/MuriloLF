# Aula 01 Importando e traduzindo dados

```python
import pandas as pd
pd.set_option('display.max_rows', 244) #SETA O MAXIMO DE LINHAS QUE APARECER√Å NO DATASET
```

```python
dados = pd.read_csv('datas/tips.csv')
```

```python
dados
```

```python
renomear = {
    'total_bill' : 'valor_da_conta',
    'tip' : 'gorjeta',
    'dessert' : 'sobremesa',
    'day' : 'dia_da_semana',
    'time' : 'hora_do_dia',
    'size' : 'total_de_pessoas'
}
```

```python
gorjetas['dia_da_semana'] = gorjetas['dia_da_semana'].map(dias)
```

```python
hora = {
    'Dinner' : 'Jantar',
    'Lunch' : 'Almo√ßo'
}
```

```python
gorjetas['hora_do_dia'] = gorjetas['hora_do_dia'].map(hora)
```

### Teste de hip√≥tese

#### H0 = A distribui√ß√£o da taxa da gorjeta √© a mesma nos dois grupos
#### H1 = A distribui√ß√£o da taxa da gorjeta √© diferente nos dois grupos

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query('sobremesa == "Sim"').porcentagem
```

```python
sem_sobremesa = gorjetas.query('sobremesa == "N√£o"').porcentagem
```

```python
r = ranksums(sobremesa, sem_sobremesa)
```

```python
print('O valor do p-value √© {}'.format(r.pvalue))
```

### Aceita hip√≥tese nula por conta do "pvalue"

### Teste de Hip√≥tese

### H1 = A distribui√ß√£o do valor da conta n√£o √© igual no s√°bado e no domingo

```python
valor_conta_domingo = gorjetas.query("dia_da_semana == 'Domingo'").valor_da_conta
```

```python
valor_conta_sabado = gorjetas.query("dia_da_semana == 'S√°bado'").valor_da_conta
```

```python
r = ranksums(valor_conta_domingo, valor_conta_sabado)
```

```python
print('O valor do p-value √© {}'.format(r.pvalue))
```

# Procurando usuarios similares

```python
import numpy as np
def distancia(a,b):
    return np.linalg.norm(a - b)
```

```python
def notas_usuario(usuario):
    notas_usuario = notas.query("usuarioId==%d" % usuario)
    notas_usuario = notas_usuario[['filmeId', 'nota']].set_index('filmeId')
    return notas_usuario
```

```python
usuario1 = notas_usuario(1)
usuario4 = notas_usuario(4)
```

```python
diferencas = usuario1.join(usuario4, lsuffix="_esquerda", rsuffix="_direita").dropna()
distancia(diferencas['nota_esquerda'],diferencas['nota_direita'])
```

```python
def distancia_usuarios(usuario_id1, usuario_id2):
    notas1 = notas_usuario(usuario_id1)
    notas2 = notas_usuario(usuario_id2)
    diferencas = notas1.join(notas2, lsuffix="_esquerda", rsuffix="_direita").dropna()
    return distancia(diferencas['nota_esquerda'],diferencas['nota_direita'])
```

```python
def distancia_usuarios(usuario_id1, usuario_id2):
    notas1 = notas_usuario(usuario_id1)
    notas2 = notas_usuario(usuario_id2)
    diferencas = notas1.join(notas2, lsuffix="_esquerda", rsuffix="_direita").dropna()
    DISTANCIA = distancia(diferencas['nota_esquerda'],diferencas['nota_direita'])
    return [usuario_id1, usuario_id2, DISTANCIA]

distancia_usuarios(1, 2)
```

```python
def distancia_todos(voce_id):
    distancias = []
    for usuario_id in notas['usuarioId'].unique():
        informacoes = distancia_usuarios(voce_id, usuario_id)
        distancias.append(informacoes)
    return distancias

distancia_todos(1)[:5]  
```

```python
def distancia_todos(voce_id):
    todos_usuarios = notas['usuarioId'].unique()
    distancias = [distancia_usuarios(voce_id, usuario_id) for usuario_id in todos_usuarios]
    distancias = pd.DataFrame(distancias, columns = ["voce", "outra_pessoa", "distancia"])
    return distancias

```
